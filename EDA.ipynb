```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# for local environment
df = pd.read_pickle('working_dataset.pkl')
df = df.astype(float)
df.head()
```

```python
# statistics of the datasets
df.describe()
```

```python
# the null values of the cleaned datasets
df.isna().sum().sum()
```

```python
# correlation coefficient
plt.figure(figsize=(16, 16))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# plot the price
plt.plot(range(len(df)), df.iloc[:, 1], color='blue', label='Real')
plt.legend()
plt.show()
```

```python
# plot the price % change
data = df.values
price_pct = [0]
for i in range(1, data.shape[0]):
    price_pct.append((data[i, 1] - data[i-1, 1])/data[i-1, 1])
plt.plot(range(len(price_pct)), price_pct, color='blue', label='Real')
plt.legend()
plt.show()
```
